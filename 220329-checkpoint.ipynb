# NbS 토픽모델링

```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import pyLDAvis.gensim_models
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from gensim.parsing.preprocessing import preprocess_string
from gensim import corpora
from gensim.models.coherencemodel import CoherenceModel
from gensim.models import LsiModel
from gensim.models import LdaModel

nltk.download('stopwords')
```

## 전처리

```python
def clean_text(d):
    pattern = r'[^a-zA-Z\s]'
    text = re.sub(pattern, '', d)
    return d

def clean_stopword(d):
    stop_words = stopwords.words('english')
    return ' '.join([w.lower() for w in d.split() if w not in stop_words and len(w) > 3])

def preprocessing(d):
    return preprocess_string(d)
```

```python
df = pd.read_excel('nbs_data.xlsx')
df
```

```python
documents = list(df['Abstract'])
print(len(documents))
documents[0]
```

```python
grouped = df.groupby(['Year'])
grouped.count()
```

```python
df_y_a = df.loc[:, ['Year', 'Abstract']]
df_a = df.loc[:, ['Abstract']]
len(df_a)
```

```python
text = str(df_y_a.loc[[0], ['Abstract']])
print(text)
print('Sentence Tokenize :',sent_tokenize(text))
```

```python
df_a.replace("", float("NaN"), inplace=True)
# df_a.isnull().values.any()
# df_a.dropna(inplace=True)
```

```python
df_a['Abstract'] = df_a['Abstract'].apply(clean_text)
df_a['Abstract']
```

```python
df_a['Abstract'] = df_a['Abstract'].apply(clean_stopword)
df_a['Abstract']
```

```python
tokenized_abs = df_a['Abstract'].apply(preprocessing)
tokenized_abs = tokenized_abs.to_list()
# tokenized_abs
```

```python
drop_abs = [index for index, sentence in enumerate(tokenized_abs) if len(sentence) <= 1]
abs_texts = np.delete(tokenized_abs, drop_abs, axis=0)
print(len(abs_texts))
```

## 토픽 모델링 (Gensim)

```python
dictionary = corpora.Dictionary(abs_texts)
corpus = [dictionary.doc2bow(text) for text in abs_texts]
print(corpus[1])
```

## 잠재 의미 분석 (LsiModel)

```python
min_topics, max_topics = 1, 15
coherence_scores = []

for num_topics in range(min_topics, max_topics):
    model = LsiModel(corpus, num_topics=num_topics, id2word=dictionary)
    coherence = CoherenceModel(model=model,
                              texts=abs_texts,
                              dictionary=dictionary)
    coherence_scores.append(coherence.get_coherence())
print(coherence_scores)
```

```python
plt.style.use('seaborn-white')
x = [int(i) for i in range(min_topics, max_topics)]
plt.figure(figsize=(10, 6))
plt.plot(x, coherence_scores)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Scores')
plt.show()
```

```python
lsi_model = LsiModel(corpus, num_topics=6, id2word=dictionary)
topics = lsi_model.print_topics(num_topics=6)
topics
```

## 잠재 드리클레 할당 (LdaModel)

```python
min_topics, max_topics = 1, 15
coherence_scores = []

for num_topics in range(min_topics, max_topics):
    model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary)
    coherence = CoherenceModel(model=model,
                              texts=abs_texts,
                              dictionary=dictionary)
    coherence_scores.append(coherence.get_coherence())
print(coherence_scores)
```

```python
plt.style.use('seaborn-white')
x = [int(i) for i in range(min_topics, max_topics)]
plt.figure(figsize=(10, 6))
plt.plot(x, coherence_scores)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Scores')
plt.show()
```

```python
lda_model = LdaModel(corpus, num_topics=9, id2word=dictionary)
topics = lda_model.print_topics(num_topics=9)
topics
```
